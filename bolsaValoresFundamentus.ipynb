# Introdução
Esse notebook foi criado com a intenção de lista as N melhores empresas listada na bolsa de valores para uma analise mais detalhada, obedecendo alguns indicadores:


1. **PL = Preço lucro**
O índice Preço/Lucro de uma ação é um índice usado para medir quão baratos ou caros os preços das ações estão, dentro de um período de comparação. É provavelmente o mais consistente indicador de ressalva quanto ao otimismo excessivo no mercado. Serve também como um marcador de problemas e de oportunidades de negócio

2. **ROE = Retorno sobre o patrimônio**
Retorno sobre o patrimônio, retorno sobre patrimônio líquido, rentabilidade, rentabilidade líquida ou retorno sobre o capital é um indicador financeiro percentual que se refere à capacidade de uma empresa em agregar valor a ela mesma utilizando os seus próprios recursos.

3. **MrgLiq = Margem Líquida**
A Margem Líquida mede a fração de cada real de vendas que resultou em lucro líquido. Corresponde ao Lucro Líquido dividido pelas vendas líquidas. As vendas líquidas correspondem àquilo que a empresa auferiu com vendas decrescido dos custos dessas vendas

4. **DivYield = Dividend yield**
Dividend Yield, abreviadamente DY, é uma expressão inglesa que traduzida literalmente significa rendimento do dividendo. É um índice criado para medir a rentabilidade dos dividendos de uma empresa em relação ao preço de suas ações.

# Metodo

Foram criandos scores para cada indicador, Ordenados por order crescente para criando um indice ordinal,com exceção do PL que é invertido, onde quanto menor o PL melhor o 'score', no final os score sao somados e gerado o score total, ordenador por orde crecentes e listados o N melhores,foram removidas empresa com baixa liquidez, Crescimento, PL e ROE negativos. O metodo foi desenvolvindo com base duas tecnicas citadas em 2 livros, o primeiro A Fórmula Mágica de Joel Greenblat, mas por se tratar de um livro sobre o mercado americano, não funcionava muito bem no mercado brasileiro, então fiz um mix com a idea do Barsi autor do livro Faça Fortuna com Ações, cita as vantagem de investir em empresa que pagam bons Dividend Yeld, entre outos diversos ensinamentos.

Ler o Html e converte dataFrame

```python
import time
import pandas as pd
import numpy as np

df = pd.read_html('resultado.html',thousands='.',decimal=',')
df = df[0]
len(df)
```

Remover crescimento 5Y negativo

```python
df = df[df['CrescRec5a'] > 0]
len(df)
```

Remover Movimentações menor que 100000 reais 

```python
df = df[df['Liq2meses'] > 100000]
len(df)
```

Remove P/L negativo

```python

df = df[df['PL'] > 0]
len(df)
```

REMOVE ROE negativo


```python
df['ROE'] = pd.to_numeric(df['ROE'].replace(',', '.', regex=True).astype(float))
df = df[df['ROE'] > 0]
len(df)
```

Criar Score P/L , ordenar o PL do menor para o menor e colocar um numero ordinal

```python
df_teste = df.sort_values(['PL'], ascending= False)
scorePL = np.arange(0, len(df_teste))
df_teste.insert(21,'score_PL',scorePL)
df = df_teste
len(df)
```

Criar Score ROE, Ordernar por ROE quanto maior melhor e colocar o score Ordinal

```python
df_teste = df.sort_values(['ROE'], ascending= True)
scoreRoe = np.arange(0, len(df_teste))
df_teste.insert(22,'score_ROE',scoreRoe)
df = df_teste
len(df)
```

```python
df_teste = df.sort_values(['MrgLiq'], ascending= True)
score_Margem_Liq = np.arange(0, len(df_teste))
df_teste.insert(23,'score_MargemLiq',score_Margem_Liq)
df = df_teste
len(df)
```

Criar score Yeld

```python
df_teste = df.sort_values(['DivYield'], ascending= True)
score_Div_Yield = np.arange(0, len(df_teste))
df_teste.insert(24,'score_DivYield',score_Div_Yield)
df = df_teste
len(df)
```

Criar columa score_total e somar todos os scores

```python
df.insert(25,'score_total', df['score_DivYield'] + df['score_MargemLiq']+ df['score_ROE'])
```

Ordernar Score total

```python
df = df.sort_values(['score_total'], ascending= False)
```

Remover papeis classe diferentes, matendo apenas o primeiro 


```python
df = df.drop_duplicates(subset=['PatrimLiq','CrescRec5a'], keep='first')
```
